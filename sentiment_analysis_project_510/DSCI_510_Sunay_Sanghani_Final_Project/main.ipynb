```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta

def download_mercury_news_headlines(out_file1: str, sample_p: bool):
    headlines = []
    for i in range(30):
        url = f'https://www.mercurynews.com/?s=&paged={i}&orderby=date'
        try:
            response = requests.get(url, timeout=30)
            if response.status_code == 200:
                soup = BeautifulSoup(response.text, 'html.parser')
                
                articles = soup.find_all('a', class_='article-title')
                #print(articles)
                for article in articles:
                    headline_tag = article.find('span', class_='dfm-title metered')
                    if headline_tag:
                        headline = headline_tag.get_text(strip=True)
                        headlines.append(headline)
                        if sample_p and len(headlines) >= 5:
                            break                
            else:
                print("Failed to retrieve webpage, status code:", response.status_code)
        except requests.exceptions.RequestException as e:
            print(f"Request failed: {e}")
    with open(out_file1, 'w+', encoding='utf-8') as file:
                    files.append(out_file1)
                    #print(headlines_today)
                    for headline in headlines:
                        file.write(f"{headline}\n")
                        
    #print(f"Successfully wrote {len(headlines)} headlines to {out_file1}")
```

```python
import os
def download_dataset1(files, output_file):
    with open(output_file, 'w') as outfile:
        for file_path in files:
            with open(file_path, 'r') as infile:
                for line in infile:
                    outfile.write(line)
                # Optionally, write a newline after each file's content
                outfile.write('\n')

    print(f"Combined {len(files)} files into {output_file}")
```

Part 2: Collecting Data from New York Times API


```python
import requests
import json
import requests
import time
from datetime import datetime, timedelta

#should take about 3-4 minutes to run

def download_headlines(api_key, num_headlines, out_file):
    sections = ["us", "world", "sports", "technology", "business", "opinion", "arts", "health", "food", "realestate", "fashion", "travel", "science", "movies", "insider"]
    total_fetched = 0
    with open(out_file, "w") as f:

        for section in sections:
            url_base = f"https://api.nytimes.com/svc/topstories/v2/{section}.json?api-key={api_key}"

            response = requests.get(url_base)
            
            if response.status_code == 200:
                #print('here')
                try:
                    articles = response.json().get('results')
                    #print(articles)
                    for article in articles:
                        if total_fetched < num_headlines:
                            title = article['title']
                            f.write(title + '\n')
                            total_fetched += 1
                        else:
                            break
                except ValueError:
                    print("Failed to decode JSON from response.")
            else:
                print(f"Failed to fetch articles, status code: {response.status_code}")
            time.sleep(12) # Wait 12 seconds before the next API call to respect the rate limit

            if total_fetched >= num_headlines:
                break

    print("Completed fetching headlines.")
```

```python
sia = SentimentIntensityAnalyzer()  # in-built model which helps train the model
```

```python
def analyze_sentiment(headlines):
    sentiments = []
    for headline in headlines:
        score = sia.polarity_scores(headline)
        sentiments.append((headline, score['compound']))
    return sentiments
```

```python
def read_headlines(file_name):
    with open(file_name, 'r', encoding='utf-8') as f:
        headlines = f.readlines()
    headlines = [line.strip() for line in headlines]
    return headlines


def summarize_sentiments(sentiment_data):
    summary = {'positive': 0, 'neutral': 0, 'negative': 0}
    for _, sentiment in sentiment_data:
        summary[sentiment] += 1
    return summary
```

```python
def categorize_sentiments(score):
        if score > 0.05:
            return 'positive'
        elif score < 0.05:
            return 'negative'
        else:
            return 'neutral'


def summarize_sentiments(sentiment_data):
    summary = {'positive': 0, 'neutral': 0, 'negative': 0}
    for _, sentiment in sentiment_data:
        #print(_)
        #print(sentiment)
        sent = categorize_sentiments(sentiment)
        summary[sent] +=1
    return summary

# Summarize sentiments
national_summary = summarize_sentiments(national_sentiments)
local_summary = summarize_sentiments(local_sentiments)
huff_summary = summarize_sentiments(huff_sentiments)
```

```python

def plot_sentiment(data, title, ax):
    x = range(len(data))  # the label locations
    width = 0.35  # the width of the bars

    rects = ax.bar(x, data.values(), width, label=title, color=['green', 'blue', 'red'])

    # Add text for labels, title, and custom x-axis tick labels, etc.
    ax.set_ylabel('Counts')
    ax.set_title(f'Sentiment Counts for {title}')
    ax.set_xticks(x)
    ax.set_xticklabels(data.keys())
    ax.legend()

    # Function to add labels on bars
    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')

    autolabel(rects)

# Create a figure and a set of subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 12), sharex=True)

# Plot for each news source
plot_sentiment(national_summary, 'National', axs[0])
plot_sentiment(local_summary, 'Local', axs[1])
plot_sentiment(huff_summary, 'HuffPost', axs[2])

fig.tight_layout()
plt.show()
```

There definetely are more negative sentiment headlines for both the national news compared to the local news. The proportion between negative to positive news of the two graphs seem to be the same.

```python
denver_sent = analyze_sentiment(denver_sent_headlines)
port_sent = analyze_sentiment(port_sent_headlines)
sj_sent = analyze_sentiment(sj_sent_headlines)
boston_sent = analyze_sentiment(boston_sent_headlines)
mia_sent = analyze_sentiment(mia_sent_headlines)
chi_sent = analyze_sentiment(chi_sent_headlines)

def get_avg_sent_score(headlines):
    avg_score = 0.0
    count = len(headlines)
    for _, score in headlines:
        avg_score += score
    return avg_score/count
```

```python
data = {
    'City': ['Denver', 'Portland', 'San Jose', 'Boston', 'Miami', 'Chicago'],
    'Sentiment Score': [get_avg_sent_score(denver_sent), get_avg_sent_score(port_sent), get_avg_sent_score(sj_sent), get_avg_sent_score(boston_sent), get_avg_sent_score(mia_sent), get_avg_sent_score(chi_sent)],  # example sentiment scores
    'Latitude': [39.7392, 45.5152, 37.3382, 42.3601, 25.7617, 41.8781],
    'Longitude': [-104.9903, -122.6784, -121.8863, -71.0589, -80.1918, -87.6298]
}

df = pd.DataFrame(data)
```

```python
import numpy as np
from prettytable import PrettyTable

def create_metric_table(scores, title):

    mean_score = np.mean(scores)
    median_score = np.median(scores)
    std_dev = np.std(scores)
    range_scores = max(scores) - min(scores)
    q75, q25 = np.percentile(scores, [75 ,25])
    iqr = q75 - q25

    table = PrettyTable()
    table.title = title
    # Add columns
    table.field_names = ["Statistic", "Value"]

    # Add rows of the table
    table.add_row(["Mean", mean_score])
    table.add_row(["Median", median_score])
    table.add_row(["Standard Deviation", std_dev])
    table.add_row(["Range", range_scores])
    table.add_row(["Interquartile Range", iqr])

    print(table)
```

```python
natl_scores = [score for _, score in national_sentiments]
local_scores = [score for _, score in local_sentiments]
huff_scores = [score for _, score in huff_sentiments]

create_metric_table(natl_scores, "National News Metrics")
create_metric_table(local_scores, "Local News Metrics")
create_metric_table(huff_scores, "Huff Post News Metrics")
```
